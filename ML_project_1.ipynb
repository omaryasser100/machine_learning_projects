```python
import pandas as pd
train=pd.read_excel('/content/drive/MyDrive/ml_assigment_2/Data_Train.xlsx')
cleaned_train=train.drop_duplicates()
cleaned_train.duplicated().sum()
cleaned_train.head(1)
```

# Data Cleaning

```python
# 1. Convert 'Date_of_Journey' from string (object) to datetime
cleaned_train['Date_of_Journey'] = pd.to_datetime(cleaned_train['Date_of_Journey'], format='%d/%m/%Y')

# 2. Extract day, month, and weekday into new columns
cleaned_train['Journey_Day'] = cleaned_train['Date_of_Journey'].dt.day
cleaned_train['Journey_Month'] = cleaned_train['Date_of_Journey'].dt.month

# 3. Drop the original date column (optional)
cleaned_train.drop('Date_of_Journey', axis=1, inplace=True)
cleaned_train.drop(['Route','Dep_Time','Arrival_Time','Additional_Info'], axis=1, inplace=True)
cleaned_train.head()
```

```python
# Extract hours and minutes as integers
def extract_duration_components(duration):
    # Initialize
    hours = 0
    minutes = 0

    # Handle missing values just in case
    if pd.isnull(duration):
        return pd.Series([0, 0])

    # Clean any leading/trailing spaces
    duration = duration.strip()

    # Extract hours
    if 'h' in duration:
        hours = int(duration.split('h')[0])
        duration = duration.split('h')[1]
    else:
        duration = duration

    # Extract minutes
    if 'm' in duration:
        minutes = int(duration.split('m')[0].strip())

    return pd.Series([hours, minutes])

# Apply to your dataframe
cleaned_train[['Duration_hours', 'Duration_minutes']] = cleaned_train['Duration'].apply(extract_duration_components)

cleaned_train.drop('Duration', axis=1, inplace=True)

```

```python
cleaned_train['Duration_in_minutes']=cleaned_train['Duration_hours']*60+cleaned_train['Duration_minutes']
cleaned_train.drop(['Duration_hours','Duration_minutes'],axis=1,inplace=True)
cleaned_train.head()
```

# Column seperation for the data pre processing

```python
nominal_columns=['Airline','Source','Destination']
ordinal_columns=['Total_Stops']
numeric_columns=['Duration_in_minutes','Journey_Day','Journey_Month']
cleaned_train.dropna(inplace=True)
cleaned_train.info()
```

# Data pre processing

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, RobustScaler
from sklearn.model_selection import train_test_split

# Split the dataset (correct order: x, y = features, target)
X = cleaned_train.drop('Price', axis=1)
y = cleaned_train['Price']

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)

# Scale numeric columns using RobustScaler
scaler = RobustScaler()
x_train[numeric_columns] = scaler.fit_transform(x_train[numeric_columns])
x_test[numeric_columns] = scaler.transform(x_test[numeric_columns])

# Encode nominal categorical columns using OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop='first')
x_train_encoded = encoder.fit_transform(x_train[nominal_columns])
x_test_encoded = encoder.transform(x_test[nominal_columns])

# The encoder returns NumPy arrays → convert to DataFrames and align column names
import pandas as pd
encoded_columns = encoder.get_feature_names_out(nominal_columns)
x_train_encoded_df = pd.DataFrame(x_train_encoded, columns=encoded_columns, index=x_train.index)
x_test_encoded_df = pd.DataFrame(x_test_encoded, columns=encoded_columns, index=x_test.index)

# Drop old nominal columns and concat encoded versions
x_train = x_train.drop(nominal_columns, axis=1).join(x_train_encoded_df)
x_test = x_test.drop(nominal_columns, axis=1).join(x_test_encoded_df)

# Encode ordinal columns using OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
x_train[ordinal_columns] = ordinal_encoder.fit_transform(x_train[ordinal_columns])
x_test[ordinal_columns] = ordinal_encoder.transform(x_test[ordinal_columns])

x_train.info()
```

# using hyper parameter tuning with optuna to acheive slight improvements in the results

```python
import optuna
import xgboost as xgb
from sklearn.metrics import r2_score
import numpy as np

def objective(trial):
    # Define hyperparameters to tune
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'random_state': 42,
        'verbosity': 0,
    }

    # Model with current trial params
    model = xgb.XGBRegressor(**params)

    # Train on log target
    model.fit(x_train, np.log(y_train))

    # Predict on test and evaluate R² on reversed log
    y_pred_log = model.predict(x_test)
    y_pred = np.exp(y_pred_log)
    r2 = r2_score(y_test, y_pred)

    return r2  # Higher is better
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)  # You can increase this later

best_params = study.best_params
print("Best Hyperparameters:", best_params)
```

```python


best_model = xgb.XGBRegressor(**best_params)
best_model.fit(x_train, np.log(y_train))


print("Final R² after tuning:", best_model.score(x_test,np.log(y_test)))
```
